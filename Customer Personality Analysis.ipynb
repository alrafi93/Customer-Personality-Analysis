```python
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=100)

plt.pie(df['Education'].value_counts(),labels= df['Education'].value_counts().index,colors = sns.color_palette('pastel')[0:5],autopct='%.0f%%') 
plt.show()
```

```python
figure(figsize=(8, 6), dpi=80)

sns.set_theme(style = 'darkgrid')
sns.countplot(x = df['Marital_Status'].to_numpy())

plt.show()
```

```python
figure(figsize=(8, 6), dpi=80)

sns.lineplot(data=df, x="Year_Birth", y="Income")
```

```python
figure(figsize=(8, 6), dpi=80)

sns.barplot(data=df, x="Marital_Status", y="Income")
```

```python
import plotly.express as px

fig = px.bar(df, x='Marital_Status', y='Number_of_Children')
fig.show()
```

```python
fig = px.scatter(df, x="Year_Birth", y="MntWines",labels={
                     "Year_Birth": "Year Birth",
                     "MntWines": "Amount Spent on Wine"                 
                 }, color = 'MntWines',title='Amount spent on wine in last 2 years by Age')
fig.show()
```

```python
fig = px.scatter(df, x="Income", y="MntMeatProducts",labels={
                     "Income": "Income",
                     "MntMeatProducts": "Amount Spent on Meat Products"                 
                 }, color = 'MntMeatProducts',title='Amount spent on meat in last 2 years by Income')
fig.show()
```
